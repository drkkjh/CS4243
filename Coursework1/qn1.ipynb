```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

def diff(img, res):
    adiff = np.mean( abs(1.0 * img - 1.0 * res))
    msdiff = np.mean((1.0 * img - 1.0*res)**2)
    return adiff, msdiff
```

Use cv2.resize() on grayscale image named 20220511_105950gl.jpg

```python
im_file = '../CS4243_2023_images_small/20220511_105950gl.jpg'
img = cv2.imread(im_file)
plt.imshow(img) 
```

```python
zoa_x0 = cv2.resize(img, None, fx = 0.25, fy = 0.25, interpolation=cv2.INTER_NEAREST)
zia_x0 = cv2.resize(zoa_x0, None, fx = 4.00, fy = 4.00, interpolation=cv2.INTER_NEAREST)

zoa_x1 = cv2.resize(img, None, fx = 0.25, fy = 0.25, interpolation=cv2.INTER_LINEAR)
zia_x1 = cv2.resize(zoa_x1, None, fx = 4.00, fy = 4.00, interpolation=cv2.INTER_LINEAR)

zoa_x2 = cv2.resize(img, None, fx = 0.25, fy = 0.25, interpolation=cv2.INTER_CUBIC)
zia_x2 = cv2.resize(zoa_x2, None, fx = 4.00, fy = 4.00, interpolation=cv2.INTER_CUBIC)

zoa_x3 = cv2.resize(img, None, fx = 0.25, fy = 0.25, interpolation=cv2.INTER_AREA)
zia_x3 = cv2.resize(zoa_x3, None, fx = 4.00, fy = 4.00, interpolation=cv2.INTER_AREA)

adiff0, _ = diff(img, zia_x0)
adiff1, _ = diff(img, zia_x1)
adiff2, _ = diff(img, zia_x2)
adiff3, _ = diff(img, zia_x3)

print(adiff0)
print(adiff1)
print(adiff2)
print(adiff3)
```

Conclusion: adiff2 < adiff1 < adiff3 < adiff0 (option (c) in Canvas)
