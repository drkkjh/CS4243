```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python

img = cv2.imread('./CS4243_2023_images_small/noisy_car5.jpg')
# or noisy_car15.jpg , for that ksize-15
print(img.shape)
filteredImg = cv2.medianBlur(img, ksize=5)

# cv2.imshow('Original image', img)
# cv2.imshow('Filtered image', filteredImg)

# cv2.waitKey()
# cv2.destroyAllWindows()
plt.figure(figsize = (10, 5))
plt.subplot(121), plt.imshow(img), plt.title('Original Image'), plt.axis('off')
plt.subplot(122), plt.imshow(filteredImg), plt.title('Filtered Image'), plt.axis('off')
```
